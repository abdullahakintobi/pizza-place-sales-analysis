# 📊 PIZZA PLACE SALES ANALYSIS 📊

### Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Configure display and style
pd.set_option('display.max_columns', None)
sns.set(style="whitegrid")
```

### Load Datasets

```python
orders_df = pd.read_csv("datasets/orders.csv")
order_details_df = pd.read_csv("datasets/order_details.csv")
pizzas_df = pd.read_csv("datasets/pizzas.csv")
pizza_types_df = pd.read_csv("datasets/pizza_types.csv", encoding='latin1')
```

### Merge Datasets

```python
order_details_pizzas = pd.merge(order_details_df, pizzas_df, on="pizza_id", how="left")
order_details_full = pd.merge(order_details_pizzas, pizza_types_df, on="pizza_type_id", how="left")
full_df = pd.merge(order_details_full, orders_df, on="order_id", how="left")
```

```python
full_df.info()
```

```python
full_df.head()
```

---

### Data Preparation

```python
full_df["datetime"] = pd.to_datetime(full_df["date"] + " " + full_df["time"])
full_df["date"] = pd.to_datetime(full_df["date"])
full_df["hour"] = full_df["datetime"].dt.hour
full_df["day_of_week"] = full_df["datetime"].dt.day_name()
full_df["month"] = full_df["datetime"].dt.month_name()
```

```python
full_df.head(3)
```

### Key Metrics Summary

```python
total_sales = (full_df["quantity"] * full_df["price"]).sum()
total_quantity = full_df["quantity"].sum()
total_orders = full_df["order_id"].nunique()
total_pizza_types = full_df["pizza_type_id"].nunique()
average_price = np.average(full_df["price"], weights=full_df["quantity"])
```

```python
summary = pd.DataFrame({
    "Metric": [
        "Total Revenue (USD)",
        "Total Quantity Sold",
        "Total Orders",
        "Total Pizza Types",
        "Average Pizza Price (USD)"
    ],
    "Value": [
        f"${total_sales:,.2f}",
        total_quantity,
        total_orders,
        total_pizza_types,
        f"${average_price:.2f}"
    ]
})

summary.set_index("Metric", inplace=True)

summary
```

### Analysis Visualization

```python
# Peak Hours of Sales
peak_hours = full_df.groupby("hour")["quantity"].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
sns.barplot(x=peak_hours.index, y=peak_hours.values)
plt.title("Peak Hours of Sales")
plt.xlabel("Hour of the Day")
plt.ylabel("Total Quantity Sold")
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
# Sales by Day of the Week
weekday_order = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
sales_by_day = full_df.groupby("day_of_week")[["quantity", "price"]].apply(lambda df: (df["quantity"] * df["price"]).sum()).reindex(weekday_order)
plt.figure(figsize=(10, 6))
sns.barplot(x=sales_by_day.index, y=sales_by_day.values)
plt.title("Total Sales by Day of the Week")
plt.xlabel("Day of the Week")
plt.ylabel("Sales (USD)")
plt.tight_layout()
plt.show()
```

```python
# Top 5 Bestselling Pizzas
top_5_pizzas = full_df.groupby("name")["quantity"].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_5_pizzas.values, y=top_5_pizzas.index)
plt.title("Top 5 Bestselling Pizzas")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Pizza Name")
plt.tight_layout()
plt.show()
```

```python
# Monthly Sales Trend
months_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]
sales_by_month = full_df.groupby("month")[["quantity", "price"]].apply(lambda df: (df["quantity"] * df["price"]).sum()).reindex(months_order)
plt.figure(figsize=(12, 6))
sns.barplot(x=sales_by_month.index, y=sales_by_month.values)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Underperforming Pizzas
low_performers = full_df.groupby("name")["quantity"].sum().sort_values().head(5)
plt.figure(figsize=(10, 6))
sns.barplot(x=low_performers.values, y=low_performers.index)
plt.title("Bottom 5 Underperforming Pizzas")
plt.xlabel("Total Quantity Sold")
plt.ylabel("Pizza Name")
plt.tight_layout()
plt.show()
```

### Insights Summary
- Total revenue exceeded $800K
- Saturday is the most profitable day
- Sales peak between 12 PM and 1 PM
- A handful of pizza types dominate sales
- Some pizzas consistently underperform and may require review
